```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
email=[]
# Send a GET request to the URL
for i in range(1,15):
    url = 'http://kamarajengg.edu.in/faculty/'+str(i)+'/'
    response = requests.get(url)
    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, 'html.parser')
    html=soup.text
    x = re.findall("[a-z]*@kamarajengg.edu.in", html)
    email.extend(x)
```

```python

print(email)
for i in email:
    print(i)
```
