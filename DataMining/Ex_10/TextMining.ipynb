```python
#Import and download necessary modules

import nltk

nltk.download('punkt')

nltk.download('stopwords')

nltk.download('wordnet')

nltk.download('omw-1.4')

nltk.download('averaged_perceptron_tagger')
```

```python
#Sentence Tokenization

from nltk.tokenize import sent_tokenize

text = '''A hacker is a person skilled in information technology who uses their technical knowledge to achieve a goal or overcome an obstacle, within a computerized system by non-standard means. Though the term hacker has become associated in popular culture with a security hacker – someone who utilizes their technical know-how of bugs or exploits to break into computer systems and access data which would otherwise be inaccessible to them – hacking can also be utilized by legitimate figures in legal situations. For example, law enforcement agencies sometimes use hacking techniques in order to collect evidence on criminals and other malicious actors. This could include using anonymity tools (such as a VPN, or the dark web) to mask their identities online, posing as criminals themselves. Likewise, covert world agencies can employ hacking techniques in the legal conduct of their work. On the other hand, hacking and cyber-attacks are used extra- and illegally by law enforcement and security agencies (conducting warrantless activities), and employed by state actors as a weapon of both legal and illegal warfare.'''

tokenized_text = sent_tokenize(text)

print(tokenized_text)
```

```python
#Word Tokenization

from nltk.tokenize import word_tokenize

tokenized_word = word_tokenize(text)

print(tokenized_word)
```

```python
#Frequency distribution of all the words in the text

from nltk.probability import FreqDist

frequency = FreqDist(tokenized_word)

print (frequency)
```

```python
frequency.most_common(3)
```

```python
# Plot the frequency distribution of all the words in the text

import matplotlib.pyplot as plt

frequency.plot(30, cumulative=False)

plt.show()
```

```python
#Stopwords in English

from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

print(stop_words)
```

```python
# Tokenization and Filtering

filtered_sent = []

for w in tokenized_text:

    if w not in tokenized_word:

        filtered_sent.append(w)

print ("Tokenized Sentence: ", tokenized_text)

print ("Filtered Sentence: ", filtered_sent)
```

```python
#Stemmming

from nltk.stem import PorterStemmer

from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()

stemmed_words = []
```

```python
for w in filtered_sent:

    stemmed_words.append(ps.stem(w))

print("Stemmed Sentence:",stemmed_words)
```

```python
#Lemmarization

from nltk.stem.wordnet import WordNetLemmatizer

lem = WordNetLemmatizer()

from nltk.stem import PorterStemmer

stem = PorterStemmer()
```

```python
word="Hacking"

print("Lemmatized Word: ", lem.lemmatize(word, "v"))

print("Stemmed Word: ", stem.stem(word))

nltk.pos_tag(tokenized_word)
```
