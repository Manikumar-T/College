```python
# Load libraries

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
```

```python
#Load the dataset

data = pd.read_csv('/mnt/F84E0CA94E0C632C/CollegeClass/SEM_6/College/DataMining/Ex_8/hierarchial_dataset.csv')

print (data.head())
```

```python
#Normalize the dataset and bring all the variables to the same scale

from sklearn.preprocessing import normalize

data_scaled = normalize(data)

data_scaled = pd.DataFrame(data_scaled, columns=data.columns)

print(data_scaled.head())
```

```python
#Fit the Model

import scipy.cluster.hierarchy as shc

model = shc.linkage(data, method='ward')
```

```python
#Draw the dendrogram to decide the number of clusters

plt.figure(figsize=(10, 7))

plt.title("Dendrograms")

dend = shc.dendrogram(model)
```
